```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from pathlib import Path
import pickle

from sklearn.compose import ColumnTransformer, make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import FunctionTransformer, OneHotEncoder

from sklearn import set_config; set_config(display='diagram')
import pymupdf
from detect_ai_content.ml_logic.for_texts.using_ml_features.TrueNetTextLogisticRegression import TrueNetTextLogisticRegression
from detect_ai_content.ml_logic.for_texts.lgbm_internal import LgbmInternal
```

```python
# doc = pymupdf.open("/Users/jeromemorissard/Downloads/poa.pdf") # AI
doc = pymupdf.open("/Users/jeromemorissard/Downloads/A Son Of The Sun.pdf") # Human

header = "Header"  # text in header
footer = "Page %i of %i"  # text in footer
texts = []
for page in doc:
    texts.append(page.get_text())

model = LgbmInternal().pretrained_model()
# model = TrueNetTextLogisticRegression().local_trained_pipeline()

df = pd.DataFrame(data=texts, columns=['text'])
results = model.predict(df)
results_all_probas = model.predict_proba(df)
results_proba = []

for i in range(len(results)):
    c = results[i]
    (p0, p1) = results_all_probas[i]
    if c == 0:
        results_proba.append(p0)
    else:
        results_proba.append(p1)

print(results_proba)
df = pd.DataFrame(data={ "generated": results, 'proba': results_proba})
print(df)

formated_result = f"{100 * np.mean(results)}% of the pages of the book are generated by AI"
print(formated_result)
```
